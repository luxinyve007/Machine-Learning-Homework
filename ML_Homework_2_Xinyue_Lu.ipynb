# Compare the two models: your first function for model selection

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
def my_first_model_selection_fun(test_size = 0.3, random_state = 0):
    """
    The function has the following major steps:
    1. Divide the data into training and testing using a single split
    2. Standardize the features and target
    3. Train each model on the training data
    4. Test each model on the testing data
    5. Plot each model on the testing data
    6. Get the accuracy of each model
    7. First print the accuracy of each model and then print the better model (with the higher accuracy)
    
    Parameters
    ----------
    test_size : the proportion for testing, 0.3 by default
    random_state : the seed used by the random number generator, 0 by default
    """
    
    # Divide the data into training and testing (using the above test_size and random_state)
    # Implement me
    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = test_size, random_state = random_state )
    
    # Standardize the features and the target
    ss = StandardScaler()
    X_train = ss.fit_transform(X_train)
    y_train = ss.fit_transform(y_train.reshape(-1, 1)).reshape(-1)
    X_test = ss.transform(X_test)
    y_test = ss.transform(y_test.reshape(-1, 1)).reshape(-1)
    
    # Declare the linear regression model
    slr = LinearRegression()
    
    # The accuracy of the two models
    scores = []
    
    for feature in [alpha, beta]:
        # Get the index of the feature
        idx = list(df.columns).index(feature)

        # Train the model on the training set
        slr.fit(X_train[:, idx].reshape(-1, 1), y_train)

        # Test the model on the testing set
        y_pred = slr.predict(X_test[:, idx].reshape(-1, 1))    

        # Plot the model on the testing set
        plt.scatter(X_test[:, idx], y_test, c='steelblue', edgecolor='white', s=70)
        plt.plot(X_test[:, idx], y_pred, color='black', lw=2)  
        plt.xlabel(feature)
        plt.ylabel(target)
        plt.show()

        # Add the accuracy of the model
        scores.append(slr.score(X_test[:, idx].reshape(-1, 1), y_test))
        
    # Get the accuracy of the model
    score_alpha, score_beta = scores
    
    # First print the accuracy of each model
    # then print the better model (with the higher accuracy)
    my_print_fun(score_alpha, score_beta)
```

```python
def my_print_fun(score_alpha, score_beta):
    """
    First print the accuracy of each model
    then print the better model (with the higher accuracy)
    
    Parameters
    ----------
    score_alpha : the accuracy of model alpha
    score_beta : the accuracy of model beta
    """
    
    # Print the accuracy
    print('Accuracy for feature ' + alpha + ': %.3f' % score_alpha)
    print('Accuracy for feature ' + beta + ': %.3f' % score_beta, end='\n\n')

    # Print the better model
    if score_alpha > score_beta:
        print('The better model is the one based on feature: ' + alpha)
    elif score_alpha < score_beta:
        print('The better model is the one based on feature: ' + beta)
    else:
        print('The two models are equally accurate')
```

# Compare the two models: your second function for model selection

As shown in the figures in secs. 7.1 and 7.2, the better model obtained from the two runs are different (LSTAT from the first and RM from the second). This is due to the random split (for training and testing) obtained from the train_test_split function. Intuitively, if we test a model using multiple splits, then the average accuracy across all the splits will be more reliable than the accuracy obtained using a single split. This is the key reason why cross validation is required for model selection.

# Compare the two models: your third function for model selection

```python
from sklearn.model_selection import KFold

def my_third_model_selection_fun(n_splits = 10, random_state = 0):
    """
    The function has the following major steps:
    1. Divide the data into training and testing using multiple splits
    2. Standardize the features and target
    3. Train each model on the training data
    4. Test each model on the testing data
    5. Get the average accuracy of each model obtained by cross validation
    6. First print the accuracy of each model and then print the better model (with the higher accuracy)
    
    Parameters
    ----------
    n_splits : the number of folds, 10 by default
    random_state : the seed used by the random number generator, 0 by default
    """
    
    # The accuracy of alpha and beta on each fold
    accs = {alpha: [], beta: []}
    
    # KFold (using the above n_splits and random_state)
    # Implement me
    kf = KFold(n_splits = n_splits, random_state = random_state)
    
    # Declare the linear regression model
    slr = LinearRegression()

    for train_idx, test_idx in kf.split(X, y):
        # Get the training and testing data using train_idx and test_idx
        # Implement me
        X_train, y_train = X.iloc[train_idx], y[train_idx]  ## X is dataframe, y is array
        X_test, y_test = X.iloc[test_idx], y[test_idx]

        # Standardize the features and the target
        # Implement me
        ss = StandardScaler()
        X_train = ss.fit_transform(X_train)
        X_test = ss.transform(X_test)
        y_train = ss.fit_transform(y_train.reshape(-1, 1)).reshape(-1)
        y_test = ss.transform(y_test.reshape(-1, 1)).reshape(-1)
        for feature in [alpha, beta]:
            # Get the index of the feature
            idx = list(df.columns).index(feature)

            # Train the model on the training set
            slr.fit(X_train[:, idx].reshape(-1, 1), y_train)
            
            # Append the accuracy
            accs[feature].append(slr.score(X_test[:, idx].reshape(-1, 1), y_test))
 
    # Get the average accuracy of each model obtained by cross validation
    # Implement me
    # Note: you do not have to use one line of code to find score_alpha or score_beta (although this can be done as a one-liner)
    # Hint: accs[alpha] / accs[beta] is the list of accuracy obtained by cross validation
    # Hint: np.mean() can be used to calculate the mean of a list
    score_alpha, score_beta = np.mean(accs[alpha]),np.mean(accs[beta])
        
    # First print the accuracy of each model
    # then print the better model (with the higher accuracy)
    my_print_fun(score_alpha, score_beta)
```

## Discussion
The results above show that, the better model selected using cross validation is LSTAT. We should be more confident about this result, for reasons mentioned in sec. 7.3. It is worth noting that while your third function is correct, it is not the best practice for data preprocessing and model selection. A much easier way to do so is using the combination of sklearn Pipeline, cross_val_score, and KFold.

# Compare the two models: your fourth function for model selection

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score

def my_fourth_model_selection_fun(n_splits = 10, random_state = 0):
    """
    The function has the following major steps:
    1. Get the average accuracy of each model obtained by cross validation
    2. First print the accuracy of each model and then print the better model (with the higher accuracy)
    
    Parameters
    ----------
    n_splits : the number of folds, 10 by default
    random_state : the seed used by the random number generator, 0 by default
    """
        
    # The pipeline with StandardScaler and LinearRegression
    # Implement me
    pipe_slr = Pipeline([('StandardScaler', StandardScaler()), ('LinearRegression', LinearRegression())])

    # Get the list of accuracy obtained by cross_val_score (using KFold, and tha above n_splits and random_state)
    # Implement me
    scores_alpha = cross_val_score(pipe_slr, 
                                   X.iloc[:, list(df.columns).index(alpha)].values.reshape(-1, 1),  ## X is dataframe should change to array
                                   y, 
                                   cv=KFold(n_splits = n_splits, random_state = random_state))
    scores_beta = cross_val_score(pipe_slr, 
                                  X.iloc[:, list(df.columns).index(beta)].values.reshape(-1, 1), 
                                  y, 
                                  cv=KFold(n_splits = n_splits, random_state = random_state))
        
    # Get the average accuracy of each model obtained by cross validation
    score_alpha, score_beta = scores_alpha.mean(), scores_beta.mean()
    
    # First print the accuracy of each model
    # then print the better model (with the higher accuracy)
    my_print_fun(score_alpha, score_beta)
```

## Run your fourth function

```python
my_fourth_model_selection_fun(10, 0)
```
